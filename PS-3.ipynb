# Question 6

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
plt.figure(figsize=(10,6))
plt.title('Us marriage divorce data', fontdict={'fontweight':'bold', 'color': 'green','fontsize': 18})
plt.plot(us_marriage_divorce_data.Year, us_marriage_divorce_data.Marriages, 'b.-', label='Marriages')
plt.plot(us_marriage_divorce_data.Year, us_marriage_divorce_data.Divorces, 'r.-', label='Divorces')
plt.xlabel('Year')
plt.ylabel('Marriages\Divorces per capita')
plt.legend()
plt.savefig('Us marriage divorce data.png', dpi=350)
plt.show()
```

# Question 7

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

US_DivorceData= pd.read_csv('us-marriages-divorces-1867-2014.csv')
US_DivorceData= US_DivorceData [
    US_DivorceData ['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = tuple(US_DivorceData ['Year'].values.tolist())
marriages = tuple(US_DivorceData ['Marriages'].values.tolist())
divorces= tuple(US_DivorceData ['Divorces'].values.tolist())

# create plot
n_groups = 3

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.45
opacity = 0.9

rects1 = plt.bar(index, marriages, bar_width,
alpha=opacity,
color='y',
label='Marriage')

rects2 = plt.bar(index + bar_width, divorces, bar_width,
alpha=opacity,
color='r',
label='Divorce')

plt.xlabel('Years')
plt.ylabel('Marriage\Divorce per capita')
plt.title('Us Marriage and Divorce data')
plt.xticks(index + bar_width, (1900, 1950, 2000))
plt.legend()

plt.tight_layout()
plt.show()
```
